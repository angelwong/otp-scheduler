```python
rests_lunch = restaurants.iloc[:, 9:16].fillna(0).astype(int)
rests_lunch['name'] = restaurants['name'] # reinsert at end - name
rests_lunch['code'] = restaurants.index # reinsert at end - code
rests_lunch = rests_lunch.to_numpy()
rests_lunch
```

```python
hospitals = sort_schedule(parse.parse_from_hospital_requests_csv('../../examples/boston/20200405/hospitals.csv'))
```

```python
hosps_lunch = hospitals.iloc[:, 12:19].fillna(0).astype(int)
hosps_lunch['name'] = hospitals['name'] # reinsert at end - name
hosps_lunch['zone'] = hospitals['zone'] # reinsert at end - zone
hosps_lunch['vip'] = False # reinsert at end - vip (assume all false for now)
hosps_lunch['priority_rank'] = hospitals['priority_rank'] # reinsert at end - rank
hosps_lunch['code'] = hospitals['code'] # reinsert at end - rank
hosps_lunch = hosps_lunch.to_numpy()
hosps_lunch
```

```python
### POC on Lunch only

from __future__ import print_function
from ortools.sat.python import cp_model

def main():    
#     # sample hospital row:
#     #   hosp[0-6] = first 7 items are capacity for Mon thru Sun
#     #   hosp[7] = hospital name
#     #   hosp[8] = hospital delivery run cluster
#     #   hosp[9] = is hospital VIP? (NOTE: assume that if hospital is vip all their orders for all days are fulfilled)
#     hosp = [
#         [60, 60, 60, 60, 60, 60, 60, 'A', '1', False],
#         [30, 30, 30, 30, 30, 30, 30, 'B', '1', False],
#         [0, 0, 100, 0, 0, 100, 0, 'C', '1', True],
#         [0, 0, 45, 0, 0, 45, 0, 'D', '1', True],
#         [12, 12, 12, 12, 12, 12, 12, 'E', '1', False],
#         [35, 0, 35, 0, 35, 0, 0, 'F', '1', False],
#         [25, 25, 25, 25, 25, 0, 0, 'G', '2', False],
#         [600, 600, 600, 600, 600, 300, 300, 'H', '2', False]
#     ]
    
#     # sample restaurant row:
#     #   rest[0-6] = first 7 items are capacity for Mon thru Sun
#     #   rest[7] = restaurant name
#     rest = [
#         [600, 600, 600, 600, 600, 600, 600, 'A'],
#         [700, 700, 700, 700, 700, 700, 700, 'B'],
#         [150, 150, 150, 150, 150, 150, 150, 'C'],
#         [200, 200, 200, 200, 200, 200, 200, 'D'],
#         [300, 300, 300, 300, 300, 300, 300, 'E']
#     ]    
    hosp = hosps_lunch
    rest = rests_lunch

    ## GLOBAL_CONFIG
    MEAL_CAPACITY = 10000
    MAX_DROPOFFS_PER_RUN = 4
    MIN_MEALS_PER_RUN = 20
    DAYS_OF_WEEK = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
    
    num_hosps = len(hosp)
    num_rest = len(rest)
    num_days = 7
    
    clusters = {}

    all_hosps = range(num_hosps)
    all_rests = range(num_rest)
    all_days = range(num_days)

    # Create delivery run clusters
    for h in range(len(hosp)):
        curr_cluster = hosp[h][8]
        if curr_cluster in clusters:
            clusters[curr_cluster].append(h)
        else:
            clusters[curr_cluster] = [h]        
    
    # Creates the model.
    model = cp_model.CpModel()

    # Creates capacity variables.
    # capacity[(d, h, r)]: on day d, hospital h is served by restaurant r
    capacity = {}
    for d in all_days:
        for h in all_hosps:
            for r in all_rests:
                capacity[(d, h, r)] = model.NewBoolVar('capacity_d%ih%ir%i' % (d, h, r))

    # A hospital can be served by at most one restaurant per day
    for d in all_days:
        for h in all_hosps:
            model.Add(sum(capacity[(d, h, r)] for r in all_rests) <= 1)
                
    # A restaurant can serve at most up to its capacity per day
    for d in all_days:
        for r in all_rests:
            model.Add(sum(hosp[h][d] * capacity[(d, h, r)] for h in all_hosps) <= rest[r][d])

    # A restaurant must serve at least MIN_MEALS_PER_RUN (50) in a given cluster or none at all
#     for d in all_days:
#         for r in all_rests:
#             for cl in clusters:
#                 model.Add(sum(hosp[h][d] * capacity[(d, h, r)] for h in clusters[cl]) >= 
#                           MIN_MEALS_PER_RUN * sum(capacity[(d, h, r)] for h in clusters[cl]))

    # A restaurant can only have at most MAX_DROPOFFS_PER_RUN (4) dropoffs per run
    for d in all_days:
        for r in all_rests:
            for cl in clusters:
                model.Add(sum(capacity[(d, h, r)] for h in clusters[cl]) <= MAX_DROPOFFS_PER_RUN)                
                
    # A hospital must have at least one order per week
#     for h in all_hosps:
#         model.Add(sum(hosp[h][d] * capacity[(d, h, r)] for d in all_days for r in all_rests) >= 0)

    # A hospital order is at most the requested capacity - we won't serve more than they ask for
    for d in all_days:
        for h in all_hosps:
            model.Add(sum(hosp[h][d] * capacity[(d, h, r)] for r in all_rests) <= hosp[h][d])

    # VIP orders must be honored
    for d in all_days:
        for h in all_hosps:
            if (hosp[h][9] == True) & (hosp[h][d] > 0):
                model.Add(sum(capacity[(d, h, r)] for r in all_rests) >= 1)
                
    # cap at meal capacity
    model.Add(sum(hosp[h][d] * capacity[(d, h, r)] for d in all_days for h in all_hosps for r in all_rests) <= MEAL_CAPACITY)
    
    # objective == currently prioritizes maxing out capacity, tbd to add prioritization
    model.Maximize(sum(hosp[h][d] * capacity[(d, h, r)] for d in all_days for h in all_hosps for r in all_rests))
    
    # Creates the solver and solve.
    solver = cp_model.CpSolver()
#       solver.Solve(model)
    solution_printer = cp_model.ObjectiveSolutionPrinter()
    status = solver.SolveWithSolutionCallback(model, solution_printer)
    if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:

        output_H = {}
        output_R = {}
        hosp_names = [h[7] for h in hosp]
        rest_codes = [r[8] for r in rest]
        
        for d in all_days:
            output_hosps = []
            for h in all_hosps:
                rest_string = ''
                for r in all_rests:
                    if solver.Value(capacity[(d, h, r)]) == 1:
                        rest_string += ''.join(map(str, [rest[r][8], ' (', hosp[h][d], ')']))
                output_hosps.append(rest_string)            
            output_H[DAYS_OF_WEEK[d]] = output_hosps
        hosp_df = pd.DataFrame(output_H, index=hosp_names) 
        display(hosp_df)
                  
#         # initialise data of lists. 
#         data = {'Mon':[100, 100, 100, 100], 'Tues':[99, 98, 95, 90]}  
#         # Creates pandas DataFrame. 
#         rest_df = pd.DataFrame(data, index=rest_codes) 
        
        for d in all_days:
            output_rests = []
            for r in all_rests:
                total_meals = 0
                for h in all_hosps:
                    if solver.Value(capacity[(d, h, r)]) == 1:
                        total_meals += hosp[h][d]
                output_rests.append(total_meals)  
            output_R[DAYS_OF_WEEK[d]] = output_rests
        rest_df = pd.DataFrame(output_R, index=rest_codes) 
        rest_df["total"] = rest_df.sum(axis=1)        
        display(rest_df)        

        # Statistics.
        print()
#         print(solver.ResponseStats())

        
if __name__ == '__main__':
    main()
```
